```python
# добавление нового столбца в датасет
df['Сумма'] = df['ВВП на душу населения'] + df['Социальная поддержка']
df.head(5)
```

```python
# удаление созданного столбца
df = df.drop(['Сумма'], axis=1)
df.head(5)
```

```python
# удаление последней строки
df = df.drop(df.shape[0]-1, axis=0)
df.tail(5)
```

```python
# вывод уникальных значений и их количества при помощи функции длины датасета
df['Страна или регион'].unique() # вывод самих значений
len(df['Страна или регион'].unique()) # их кол-ва
```

```python
# подсчёт количества значений (по странам)
df['Страна или регион'].value_counts()
# видим, что в датасете все страны уникальны :)
```

```python
# группировка данных
df['Баллы_new'] = round(df['Баллы'])
df.groupby('Баллы_new').count() # обычный подсчёт кол-ва значений
```

```python
# группировка с подсчётом суммы по группам
df.groupby('Баллы_new').sum()
```

```python
# группировка с агрегацией сразу по нескольким функциям
df_agg = df.groupby('Баллы_new').agg({
    'Баллы_new': 'count',
    'Баллы_new': 'sum',
    'Баллы_new': 'mean',
    'Баллы_new': 'median'
})
df_agg
```

```python
# подготовка и создание сводной таблицы
df['ВВП_new'] = round(df['ВВП на душу населения'])
pd.pivot_table(df,
               index=['Баллы_new'],
               columns=['ВВП_new'],
               values='Социальная поддержка',
               aggfunc='mean')
```

```python
# сортировка данных (по умолчанию - по возрастанию)
df.sort_values(by='ВВП на душу населения').head()
```

```python
# сортировка данных (по убыванию, добавляем параметр ascending)
df.sort_values(by='ВВП на душу населения', ascending=False).head()
```
